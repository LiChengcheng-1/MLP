```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pandas.read_excel('E:/lesson/MLP/MLP_Test Task/data/12_30 bis 13_30 Uhr Umf + Vega.xlsx')
df.info()
df.head(10)
```

```python
#according to the info assign the index
df.columns = df.iloc[7]
df.info()
```

```python
df.describe()
```

```python
#checking the missing value
missing = df.columns[df.isnull().any() ].tolist()
df[missing].isnull().sum()
```

```python
#delete the missing value
selected_df = df.dropna()
#replace the #-1 with 0
selected_df = selected_df[selected_df["app1_q [l/s]"] != 0].replace("#-1",0)
selected_df 
```

```python
selected_df =pandas.concat([selected_df.loc[8:,"app1_q [l/s]"],selected_df.loc[8:,"p1_v1 [m/s]":"p1_v16 [m/s]"]],axis=1)
selected_df
```

```python
# check if linear correleration
num_cols=[
"p1_v1 [m/s]","p1_v2 [m/s]","p1_v3 [m/s]","p1_v4 [m/s]","p1_v5 [m/s]","p1_v6 [m/s]","p1_v7 [m/s]"]
for col in num_cols:
    warnings.filterwarnings("ignore")
    sns.pairplot(selected_df[[col]+['app1_q [l/s]']])
```

```python
#draw the heat map
correlation =selected_df.corr()
fig = plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot = True,linewidths=1)
plt.show()
```

```python
#choose the most relevant features
correlation = selected_df.corr().drop(["app1_q [l/s]"])["app1_q [l/s]"]
need_remove = []
for index in correlation.index:
    if correlation[index]< 0:
        need_remove.append(index)
corr_selected_df = selected_df.drop(need_remove,axis=1)
corr_selected_df
```
