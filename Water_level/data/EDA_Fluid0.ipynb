```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pandas.read_csv('Fluid0.csv')
df.info()
df.describe()
```

```python
#checking the missing value
missing = df.columns[df.isnull().any() ].tolist()
df[missing].isnull().sum()
```

```python
# check if linear correleration
num_cols=[
"Level Sensor-1","Level Sensor-2","Level Sensor-3"]
for col in num_cols:
    warnings.filterwarnings("ignore")
    sns.pairplot(df[[col]+[' Fluid Flow']])
```

```python
#draw the heat map
correlation =df.corr()
fig = plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot = True,linewidths=1)
plt.show()
```
